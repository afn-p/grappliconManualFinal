## I. Rank and Points

Points are used as a measure of one's experience and assists with offering competitors the opportunity to match against those of similar abilities; however, competitors can still duel against individuals who may have dramatically higher level of experience. There are four types of records associated with every competitor: _Accumulated Points_, _Duel Points_, _Tournament Points_, and _Season Points_.

### Accumulated Points

This refers to the total amount of points that a competitor has earned throughout their time with Grapplicon. It is used to help match competitors with similar levels of experience for duels. Moreover, accumulated points will be used to promote individuals to the next belt rank. 

Take a look at the table below to see the rank categories and the associated points range for each level of experience.

Belt Rank | Accumulated Points
-|-
White | 0 to 4800
Blue | 5000 to 9800
Purple | 10000 to 14800
Brown | 15000 to 19800
Black | 20000

### Duel Points

Duel Points refers to the total amount of points that a competitor has earned from their first 20 Duels in the season.

This record will restart to **ZERO** once the season has concluded.

The purpose of tracking duel points is that it will be used to determine who qualifies for particular tournaments occurring at the end of every season.

### Tournament Points

Tournament points refers to the total amount of points that a competitor has earned only at the tournament.

This record will restart to **ZERO** once the tournament has concluded.

This excludes Duel Points.

### Season Points

Season points refers to the total amount of points that a competitor has earned for the current season (Duel Points Plus Any Tournament Points)

This record will restart to **ZERO** once the season has concluded. 

### ***Note: season points are only recorded for those who are registered in the _Circuit_***

## II. Circuit

The _Circuit_ is basically a series of _Leaderboards_ for which participants must compete to qualify in order to progress towards becoming world champion.

The general structure of the Circuit involves the following:

Stage | Operation
-|-
Duels | Based on the *initial 20 matches* of the season, participants compete for the best record of their registered state to qualify for the _World Championships.
Trial Tournaments | Participants compete to *achieve first place* at the _Trial Tournament_ to qualify for the _World Championships_.
World Championships | Qualifiers compete to become world champion.

Participants are required to have **AT LEAST 20 MATCHES** in the season in order to even be considered for Trial or World Championships.

### ***Note: the Circuit will primarily consist of professional level of _Black_ belt rank holders; however, with an application and sponsor, competitors of lower belt ranks may compete in the Circuit as well.***

## IV. Match Settings

- Submission-only, where competitors are awarded based on how fast they manage to submit their opponents.

- 5 minute matches - a maximum of 300 seconds per match.

- Each competitor in a match will start off with 300 points: 300 points for a 300 second match.

- As the match is playing out, competitors will lose points at a rate of _1 Point Per Second_ until the match ends.

- If the match ends by submission, the winner will receive points based on how much time is left.

- Quitting of any form is considered a _Tap_, so long as the reason is not due to a violation of safety rules.

- If the time runs out, then it is a stalemate: no one is victorious in the match, and thus no one won any points.


## V. Duel Ruleset

- Duels are conducted at authorized competition sites.

- Any one competitor can only have a maximum of one duel per day.

- Only one duel can be scheduled per competitor at a time.

- Cancellations are only permissible 48 hours, or more, prior to the scheduled duel.

- Cancellation of duels within 48 hours will result in an immediate disqualification.

### Disqualification Due to No-Show or Repeating Cancellations

_5 Points Per Hour_ is the *Rate of Deduction* starting at the beginning of the 48 hour deadline prior to the scheduled match.

A _No-Show_, where the competitor has not made any cancellation request and did not show up to their scheduled match, will result in the following:
- deduction of 500 points from both the _Season Points_ and _Accumulated Points_ of the participant.
- 1 month suspension.

2 instances of _No-Show_ in the season will result in an elimination from the _Circuit_ and a 1 year suspension.

3 cancellations will result in a deduction of 500 ponts from both the _Season Points_ and _Accumulated Points_ of the participant.

### Matches

- A series of rounds of will ensue where participants compete for a spot in the next round.

- Available spots are on the _Leaderboard_, which determines who may advance to the next round.

- Each round will consist of a series of _Matches_.

- _Matches_ will continue until all available spots on the leaderboard are occupied or when there are no more available spots for participants to take.

### Leaderboards

- The leaderboard only consists of qualified competitors.

- Leaderboards for each round are ordered based on competitors' _Performance Record_.

- Qualifying is done by the following:
1. Submitting an opponoent in the round and having a submission score that allows one to stay on the leaderboard. Participants can be kicked off the leaderboard if they are pushed further down the ordered list.

2. Having a _Bye_: if a round consists of an odd amount of competitors, and a participant is left without an opponent, the competitor with the best _Performance Record_ will immediately proceed to the next round. The leaderboard will still be ordered based on current _Performance Records_, including the participant with the _Bye_.

- Participants who have not made it on the leaderboard are ultimately eliminated from the tournament.

- Each leaderboard of the next round will have a limited amount of available spots that is at least half of the total numnber of participants competing in the current round.

- Leaderboards will decrease in available spots for the next round whenever no one manages to secure a spot: in other words, after everyone competes in their current match, the leaderboard decreases in available spots when _NOT_ one competitor manages to win. The following demonstrates the amount of spots that are eliminated whenever a total stalemate occurs in the round.

### Performance Record

Competitors are placed on the leaderboard according to their performance record as indicated by the following:

Priority | Record
-|-
1 | Total Amount of Rounds Competed for Prior Leaderboard Spot
2 | Season Points
3 | Duel Points
4 | Most points earned from any one match in the Season, including at the tournament.

- Competitor A only has 3 matches

### Champion Title Hierarchy

Only one competitor will ultimately remain, who will be awarded with a title of champion.

The performance record of the finalist will determine the champion title that they will receive.

-|-

Supreme Champion | Earned a Season Points total of at least 4800, held the highest amount of Season Points, and submitted _at least the second best performance record holder_ at the World Championships (if champion had the _second best performance record_, they will need to have ***WON*** against the _best performance record holder_ at the World Championships).

## VII. Extreme Circumstances

It is possible for no one to advance to the next round where the leaderboard has decreased so much to only have one spot left.

### If Leaderboard Is _EVEN_

If the leaderboard is even and no one advances due to consistent stalemates every round, then the ***TOP TWO*** _best performance record holders_ will face eachother as the final match of the tournament.

The winner of this match will be crowned the _Undisputed Champion_.

If a stalemate occurs, then the _best performance record holder_ will be crowned _Champion_.

### If Leadboard Is _ODD_

If the leaderboard is odd, and thus the _best performance record holder_ moved on to the next leaderboard, such that in the event where no one advances due to consistent stalemates, then the ***TOP TWO*** _best performance record holders_ will face eachother for the final match of the tournament.

The winner of this match will be crowned the _Undisputed Champion_.

If a stalemate occurs, then the _best performance record holder_ will be crowned _Champion_.

The victim - participant who was violated - will receive points depending on the time when the referee ended the match.

- The referee stopped the match because of this violation.
- _Competitor A_, the violator, ends up losing the match because of breaking the rules - diqualified.
