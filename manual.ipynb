# GRAPPLICON MANUAL

## I. Universal Match Settings

- Submission-only, where competitors are awarded based on how fast they manage to submit their opponents.

- 5 minute matches - a maximum of 300 seconds per match.

- Each competitor in a match will start off with 300 points: 300 points for a 300 second match.

- As the match is playing out, competitors will lose points at a rate of _1 Point Per Second_ until the match ends.

- If the match ends by submission, the winner will receive points based on how much time is left.

- Quitting, or Stoppoing, of any way is considered a _Tap_, so long as the reason is not due to a violation of safety rules.

- If the time runs out, then it is a stalemate: no one is victorious in the match, and thus no one won any points.

## III. Rank and Points

Points are used as a measure of one's experience and assists with offering competitors the opportunity to match against those of similar abilities; however, competitors can still duel against individuals who may have dramatically higher level of experience. There are four types of records associated with every competitor: _Accumulated Points_, _Duel Points_, _Tournament Points_, and _Season Points_.

### Accumulated Points

This refers to the total amount of points that a competitor has earned throughout their time with Grapplicon. It is used to help match competitors of similar levels of experience for duels. Moreover, accumulated points will be used to promote individuals to the next belt rank. 

Take a look at the table below to see the rank categories and the associated points range for each level of experience.

Belt Rank | Accumulated Points
-|-
White | 0 to 4500
Blue | 5000 to 9500
Purple | 10000 to 14500
Brown | 15000 to 19500
Black | 20000

### Duel Points

Duel Points refers to the total amount of points that a competitor has earned from their first 10 Duels in the season.

The purpose of tracking duel points is that it will be used to determine who qualifies for particular tournaments occurring at the end of every season.

### Trial Points

These points are primarily used during the Trials.

However, all the points earned from the _Trials_ will be added to one's _Accumulated Points_.

Again, _Trial Points_ **WILL NOT** be added to _Season Points_.

Trial points exclude Duel Points in calculations.

### Championship Points

All the points earned from the _Championships_ will be added to both one's _Accumulated Points_ and _Season Points_.

### Season Points

Season points refers to the total amount of points that a competitor has earned in the current season (_Duel Points_ and points earned at _Championships_)

## **Note: season points are only recorded for those who are registered in the _Circuit_**

## IV. Performance Record

Competitors are placed on the leaderboard according to their _Performance Record_ as indicated by the following:

Priority | Record
-|-
1 | The Total Amount of Rounds **Completed** for Next Leaderboard Spot (Lesser is Better)
2 | The Amount of Stalemates Incurred During Duels in the Current Season (Lesser is Better)
3 | Season Points, excluding Trial Points (More is Better)
4 | Duel Points (More is Better)
5 | The Most points earned from any one match in the Season, including at the tournament (More is Better)

If two competitors are equal in one priority, then the next priority will determine who will advance to a higher position on the _Leaderboard_ or be eliminated from the tournament.

As for priority one, one must actually win to be included on the next leaderboard: not earning a submission win essentially means that one is eliminated.


## V. Circuit

The _Circuit_ is basically a series of _Leaderboards_ for which participants must qualify in order to progress towards becoming champion.

Importantly, the Circuit will primarily consist of professional level _Black Belt_ rank holders; however, with an application and sponsor, competitors of lower belt ranks may compete in the Circuit as well.

Additionlly, participants will need to complete **10 Matches** during the _Duels_ if they want to ever be considered for entry into _Trials_ or _Championships_ at end of the season: these matches will be their **Initial** 10 matches of the season. Any matches after their initial 10 matches will not count. A participant's **10 Matches** will determine the following:
- earn them entry into the Championships as a _Duels Qualifier_.
- allow them to participate in the Trials, where they will need to compete for entry in to the Championships as a _Trials Qualifier_.
- if neither of the prior two possibilities occurred, the participant will be eliminated from the Circuit.

Essentially, participants will go through a series of _Leaderboards_ until only one competitor is left to be crowned champion.

Again, _Leaderboards_ are **lists** of participants who are allowed to advance in the tournament. Participants will compete for a spot on the _Leaderboards_ in order to move forward in the event. **Failure to do so would result in elimination from the tournament**.

## **Note: to earn points for the Circuit, one must compete within their registered weight division. Any submission victories completed outside of one's weight division will only be counted towards their _Accumulated Points_.**

## VI. Duels

Duels are only conducted at **Authorized Competition Sites**. 

Moreover, Duels will be available throughout the **Entire Year**.

However, during the season of the Circuit, Duels will also be used as a means of determining which registered participants can proceed to the Championships.

### General Rules

Any one competitor will only have a maximum of **One Duel Per Day**.

Only one Duel can be scheduled per participant at a time.

A participant may only compete against the same person **Once Per Season** in the Duels.

### Costs of Registering for a Duel Match

An online _Duel Registration Fee_ is split between both competitors, and paid upon completion of their match (or alternative occurrence).

Each Duel will cost this _Duel Registration Fee_.

Failure to pay the _Duel Registration Fee_, of any Duel match, will result in an immediate elimination from the Circuit.

**Note: this fee does not include any costs that are charged from the _authorized competition site_ for hosting the duel.**

### Duel Invitations

Participants will send _Duel Invitations_ to opponents.

- Opponents will need to respond to requests **Within 48 hours**.
- _Denying_ a request does not remove points from a participant's record.
- However, _Denying_ **Three** consecutive requests, and not having any pending _Duel Invitations_ sent to other participants, will result in a **1 Month** suspension.
- Similarly, _Senders_ will be suspended for **1 Month** if they have consecutively rescinded **Three** _Duel Invitations_.
- _Senders_ will need to rescind their _Duel Invitation_ **Within 48 hours** of sending the request.

Cancellations are only permissible **48 Hours, or More**, prior to the scheduled _Duel Date_.
- Cancellations may be initiated by either of the two involved parties competing against one another.
- The total cost of online registering for a _Duel_ will be completely covered by the participant who _Cancelled_, plus a _Cancellation Fee_.
- However, participants may petition to address an emergency as reason for the _Cancellation_, possibly removing the burden of the _Cancellation Fee_.
- Cancellation of duels **Within 48 hours** will result in an immediate _Season Points_ deduction.

### Additional Rules to No-Show or Repeating Cancellations of Scheduled Duel Matches.

_5 Points Per Hour_ is the *Rate of Deduction* starting at the beginning of the 48 hour deadline duration prior to a scheduled Duel match. And thus, the sooner one cancels within the 48 hour duration, the less points that they will lose.

A _No-Show_, where the competitor has not made any cancellation requests and did not show up to their scheduled match, will result in the following:
- deduction of 500 points from both the _Season Points_ and _Accumulated Points_ of the participant.
- 1 month suspension.

**2 instances of _No-Show_ in the season will result in an elimination from the _Circuit_ and a 1 year suspension.**

**3 cancellations will result in a deduction of 500 ponts from both the _Season Points_ and _Accumulated Points_ of the participant.**

If neither competitor shows up for their match, both competitors are penalized in similar fashion.

### Duel to Advance in Circuit

If registered in the Circuit, a participant's **Initial 10 Matches** of the season will count towards their _Season Score_ to be used for possible admittance into either the _Trials_ or directly to _Championships_.

Points that are earned after a competitor's **Initial 10 Matches** will not count towards their _Season Score_.

Again, a participant's _Performance Record_ will determine whether they will be one of the following:
- Duels Qualifier, who will immediately gain entry to participate at the _Championships_.
- Trials Qualifier, who will have a chance to continue competing for entry into the _Championships_.

To become a _Duels Qualifier_, registered participants in the Circuit will need to have a _Performance Record_ that is high in quality enough to earn a spot on _Leaderboard D_.

Participants compete against each other for a spot on _Leaderboard D_, where the _Performance Record_ of the highest value will be positioned at the top of the list, then ordering down other participants accordingly.

Note that earning a spot on the leaderboard does not guarantee that it will be kept: the list is ordered based on the quality of one's _Performance Record_, and thus a participant can be bumped down - or ultimately **Bumped Off** the leaderboard - if another competitor manages to be positioned higher on the leaderboard.

## VII. Trials

Trials provide a final chance for registered Circuit participants to compete for admission into the _Championships_.

**However, _Duels Qualifiers_ will _NOT_ be allowed to compete at the Trials.**

### General Porcedure

Much like how participants must compete in the Duels to earn a spot on the initial Leaderboard - _Leaderboard Q_ - of the Trials, another leaderboard is set to further eliminate participants from the Circuit.

Next, those who have made the cut to compete in the Trials will be required to earn a spot on _Leaderboard T_ if they desire to participate at the _Championships_.

Participants of the Trials will have to complete **5 Matches**, where one's _Trial Points_ will determine their position on _Leaderboard T_.

In other words, _Leaderboard T_ will not take into account one's _Performance Record_, or _Duel Points_, while ordering participants on the list. The _Trial Points_ are the only units of measure in determining where participants will be placed on the leaderboard.

Note that not earning any submission victories during the Trials will immediately elimintate one from the Circuit: submission victories are necessary to improve one's _Trial Points_ score, and thus improve one's position on _Leaderbaord T_.

Those who have managed to earn a spot on _Leaderboard T_ will be immediately permitted to compete at the _Championships_ for their weight division.

Again, one may not switch weight divisions during the Circuit: a participant may only earn points to improve their _Performance Record_ and compete in the Trials, or Championships, for the weight division that they have registered for at the beginning of the season.

## VIII. Championships

### General Procedure

Participants of the _Championships_ will consist of _Duels_ and _Trials Qualifiers_.

Again, participants must finish their initial **10 Matches** in the _Duels_ to even be considered for the following:

- A participant's initial 10 matches from the _Duels_ will be used to determine if they are worthy to compete at the _Trials_ in order to earn entry into the _Championships_.

- A participant's initial 10 matches from the _Duels_ will be used to determine if they are a _Duels Qualifier_, who will immediately have entry to participate at the _Championships_.

Whereas the _Trials_ only proceeds to have a single leaderboard for participants to compete for at the tournament, the _Championships_ will run through a **SERIES** of leaderboards.

Each leaderboard of the _Championships_, including _Leaderbaord W_ - the list that consists of both Duel and Trial Qualifiers, will order participants based on their _Performance Records_.

### Leaderboards

Again, qualifiers will be ordered based on their performance records, listing the best record at the top of the _Leaderboard_ and going down the list accordingly.

It is a series of _Leaderboards_, and thus will depenedent on the initial amount of qualified participants listed on the first leaderboard - _Leaderboard W_.

An upcoming _Leaderboard_ will have varying amounts of available spots depending on the total amount of participants on the list prior; moreover, each successive _Leaderboard_ will be at least half the amount of available spots as the one prior to it.

The better one's record is, the higher they will be on the leaderboard, and thus they will secure an opportunity to continue in the tournament.

Again, it is possible to be _Bumped Off_ the leaderboard if another partiicpant scores higher, so placing oneself higher on the list will lower the chance of being eliminated from the Circuit.

Not earning a spot on any leaderboard will ultimately result in elimintation from the tournament.

Important leaderboards of the Circuit are explained in the following, some of which reviews information of prior sections:

Leaderboard | Significance
-|-
Leaderboard D | This list consists of Duels Qualifiers, who will immediately proceed to the Championships due to their acceptable performance record from the Duels.
Leaderboard Q | This list consists of participants who are allowed to compete at the Trials.
Leaderboard T | This list consists of Trials Qualifiers, who managed to earn their entry into the Championships by their performance at the Trials.
Leaderboard W | The list of participants combining participants from both _Leaderboad D_ and _Leaderboard T_.
Leaderboard C | The participant who earns the last spot on the _Leaderboard C_ will be crowned Champion.

### Rounds

Each leaderboard of a tournament will have a series of rounds, giving participants a chance to compete for a spot on the **Next Leaderboard**.

For instance, after the _Current_ leaderboard in the tournament has been ordered based on performance records, a series of rounds will ensue to give competitors a chance to earn a spot on the _Next_ leaderboard.

When a all the matches have been complete for a round, another will start so long as there are still available spots on the _Next_ leaderboard.

Again, a series of _Rounds_ will occur until all the available spots are taken for the _Next Leaderboard_.

**Note: victories earned in earlier rounds are MORE VALUABLE than victories earned in later rounds for the upcoming leaderboard; for example, _Competitor A_ managed to secure a fast victory in _Round Three_, but _Competitor B_ would be positioned higher on the _Next_ leaderboard if their victory was secured in _Round One_ (Refer back to the Performance Record Rules).**

### Matches

Matches will be executed by order of performance record, where the better one's performance record is, the later they will compete in the round.

***See _Extreme Circumstances: Competitor Will Not Continue_ section to read about circumstances that involve participants who have quit during the tournament***

Moreover, matches are assigned based on the amount of times a participant has competed against another competitor: it is intended that participants will need to compete against everyone before having rematches.

### Final Leaderboard

Each leaderboard consistently becomes fewer in available spots as the tournament progresses, leading to the final leaderboard - _Leaderboard C_.

The finals traditionally allows for only one person to become _Champion_; essentially, this tradition is upheld by only allowing one spot available on _Leaderboard C_.

The prior leaderboard before the finals, will only consist of 2 spots available to participants, and thus the winner between the two finalists will advance.

### Champion Title Hierarchy

The champion of the tournament will be given an appropriate champion title in recogntion of their performance record.

The performance record of the finalist will determine the champion title that they will receive.

Title | Requirements
-|-
Champion | Held the highest amount of Season Points.
Undisputed Champion | Submitted _at least the second best performance record holder_ at the World Championships (if the champion had the _second best performance record_, they will need to have ***WON*** against the _best performance record holder_ at the Championships).
Supreme Champion | Earned a Season Points total of at least 2500, held the highest amount of Season Points, and submitted _at least the second best performance record holder_ at the Championships (if the champion had the _second best performance record_, they will need to have ***WON*** against the _best performance record holder_ at the World Championships).

## X. Extreme Circumstances: Consistent Stalemates

### 1. When Rounds Conclude Without Any Winnners

Understand that as available spots become less with each successive leaderboard in the tournament, leaderboards will **ALSO** decrease in available spots in the event that no one manages to win in a round - ***STALEMATE ROUND***.

In other words, after everyone competes in the current round, the _upcoming leaderboard_ decreases in available spots when **NOT** one competitor managed to secure a victory.

The _Rate of Decrease_ for spots on the _next leaderboard_ is ***n***, where ***n*** is the amount that is left over after dividing the _initially quantity of participants in the current leaderboard_ by 10, as demonstrated in the following table:

Maximum number of participants on the current leaderboard | Amount of spots eliminated from the upcoming leaderboard
-|-
10 | 1
20 | 2
30 | 3
40 | 4
50 | 5
n(10) | n

The _Initial Quantity_ of the current leaderboard is used for calculation, since the _actual_ quantity may change due quitting or disqualification.

***NOTE: the limit to how many spots that the next leaderboard can reduce to is **2**; for example, if the _rate of decrease is 3_ and there are only 3 spots left at the time of a _stalemate round_, only one spot is eliminated from the _next leaderboard_.

#### ***THE AMOUNT OF SPOTS ON A LEADERBOARD CAN NEVER BE LESS THAN TWO***

### 2. No One Advances to the Next Leaderboard

***It is possible for no one to advance to the next leaderboard where the leaderboard has decreased so much to only have one spot left.***

#### If Leaderboard Is _EVEN_

If the current leaderboard was initially even in the amount of participants, and no one advances due to _Consistent Stalemate Rounds_, then the ***TOP TWO*** _best performance record holders_ will face eachother for the final match of the tournament.

The winner of this match will be crowned the _Undisputed Champion_.

***If a stalemate occurs in this final match, then the _best performance record holder_ will be crowned _Champion_.***

#### If Leadboard Is _ODD_

If the leaderboard was initially odd in the amount of participants, and thus the _best performance record holder_ moved on to the next leaderboard - such that in the event where no one advances due to _consistent stalemate rounds_ - then the ***TOP TWO*** _best performance record holders_ will face eachother as the final match of the tournament.

The winner of this match will be crowned the _Undisputed Champion_.

***If a stalemate occurs in this final match, then the _best performance record holder_ will be crowned _Champion_.***

## IX. Extreme Circumstances: Competitor Will Not Continue

In the event that a competitor is unable, or refuses, to continue in the tournament, the following operation will occur - given that matches are executed based on performance records:

- if the current amount of participants was even, then the competitor with the best performance record will move on to the next leaderboard.

- if the current amount of participants was odd, then the competitor with the best performance record will be taken from the next leaderboard to compete in the current round.

## XI. Extreme Circumstances: Teammates

Teammates are those who represent the same gym or association.

Teammates are not allowed to compete against one another.

For every weight division, only one participant may represent a gym, or association, at a tournament - Trial Tournaments or World Championships.

***If this rule is violated, then all registered members from the gym, or association, will be supsended for 1 year.***

## XIV. Disqualification Due to Unacceptable Behaviors

If the match ends in a Disqualification due to a rule violation, then the _Violator_ will lose the match.

The _Victim_ - participant who was violated - will receive points based on the time when the referee ended the match.

For example: 

- 30 seconds have passed before _Competitor A_ did an illegal choke slam on opponent, _Competitor B_.
- The referee stopped the match because of this violation.
- _Competitor A_, the violator, ends up losing the match because of breaking the rules - diqualified.
- _Competitor B_, the one who was violated, wins the match and receives 270 points as appropriate compensation.

Only one warning may be given for the following _Prohibited Actions_: 1, 2, 3, 4, 5, 6, 7, and 8.

Otherwise, any violation will result in an immediate disqualification of the current match.

**Two _Duel_ match disqualifications will result in an elimination from the Circuit and 1 year suspension.**

**Two match disqualifications at a the _Trials_ or _Championships_ will result in an elimination from the event and 1 year suspension.**
