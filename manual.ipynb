## I. Rank and Points

Points are used as a measure of one's experience and assists with offering competitors the opportunity to match against those of similar abilities; however, competitors can still duel against individuals who may have dramatically higher level of experience. There are two records of points associated with every competitor: _Accumulated Points_ and _Season Points_.

### Accumulated Points

This refers to the total amount of points that a competitor has earned throughout their time with Grapplicon. It is used to help match competitors with similar levels of experience for duels. Moreover, accumulated points will be used to promote individuals to the next belt rank. 

Take a look at the table below to see the rank categories and the associated points range for each level of experience.

Belt Rank | Accumulated Points
-|-
White | 0 to 4800
Blue | 5000 to 9800
Purple | 10000 to 14800
Brown | 15000 to 19800
Black | 20000

### Season Points

Season points refers to the total amount of points that a competitor has earned for the current season. 

This record will restart to **zero** once the season has concluded. 

The purpose of tracking season points is that it will be used to determine who qualifies for particular tournaments occurring at the end of every season.

Note: season points are only recorded for those who are registered for the ***Circuit***.


## II. Circuit

Every season concludes with two tournaments - _Qualifier Tournaments_ and _World Championships_ - to determine a champion for the season.

_World Championship_: the final tournament of the season to determine a champion for every weight class.

_Qualifier Tournament_: winners of this tournament will qualify to compete at the world championship for their weight class.

Participants are required to have **AT LEAST 20 MATCHES** in the season in order to even be considered for any Circuit Tournaments.

Note: the Circuit will primarily consist of professional level of _Black_ belt rank holders; however, with an application and sponsor, competitors of lower belt ranks may compete in the Circuit as well.

### Qualifying for World Championships

To qualify, one of the following requirements must be fulfilled:
1. earn the highest amount of _Season Points_ in the state registered on your profile.
2. achieve first place at the _Qualifier Tournament_.

Note: if a competitor has earned the highest _Season Points_ for their state, and thus qualifies for the _World Championship_, then they are not allowed to compete at the _Qualifier Tournament_.

### Qualifier Tournament

The _Qualifier Tournament_ is there to provide the opporunity to compete at the _World Championship_.

It is open to those who have not qualified by having the highest season points in their state or do not have a state for which they compete under (only occurs if not enough registered participants, or not enough competition sites, are available in their area).

Only 50 spots are available for registration.

Participants must compete for a spot on the _Leaderboard_ to register, which is list of 50 spots that is ordered by individuals' _Season Points_.

In other words, only the top 50 highest _Season Points_ record holders will be allowed to register and compete at the qualifiers.

Again, a participant's _Season Points_ will only reflect their first 20 matches of the season and they must be competing in the _Black_ belt level rank.

## IV. Match Ruleset

- Submission-only, where competitors are awarded based on how fast they manage to submit their opponents.

- 5 minute matches - a maximum of 300 seconds per match.

- Each competitor in a match will start off with 300 points: 300 points for a 300 second match.

- As the match is playing out, competitors will lose points at a rate of _1 Point Per Second_ until the match ends.

- If the match ends by submission, the winner will receive points based on how much time is left.

- Quitting of any form is considered a _Tap_, so long as the reason is not due to a violation of safety rules.

- If the time runs out, then it is a stalemate: no one is victorious in the match, and thus no one won any points.


## V. Duel Ruleset

- Duels are conducted at authorized competition sites.

- Any one competitor can only have a maximum of one duel per day.

- Only one duel can be scheduled per competitor at a time.

- Cancellations are only permissible 48 hours, or more, prior to the scheduled duel.

- Cancellations of duels within 48 hours will result in a forfeiture.

## VI. Tournament Ruleset

Tournaments will initially consist of a full roster.

In the end, only one competitor remains after a consistent process of elimination.

The remaining competitor will be crowned champion.

The winner of the champion will be given an appropriate title in recogntion of their performance record.

### Matches

- A series of rounds of will ensue where participants compete for a spot in the next round.

- Available spots are on the _Leaderboard_, which determines who may advance to the next round.

- Each round will consist of a series of _Matches_.

- _Matches_ will continue until all available spots on the leaderboard are occupied or when there are no more available spots for participants to take.

### Leaderboards

- The leaderboard only consists of qualified competitors.

- Leaderboards for each round are ordered based on competitors' _Performance Record_.

- Qualifying is done by the following:
1. Submitting an opponoent in the round and having a submission score that allows one to stay on the leaderboard. Participants can be kicked off the leaderboard if they are pushed further down the ordered list.

2. Having a _Bye_: if a round consists of an odd amount of competitors, and a participant is left without an opponent, the competitor with the best _Performance Record_ will immediately proceed to the next round. The leaderboard will still be ordered based on current _Performance Records_, including the participant with the _Bye_.

- Participants who have not made it on the leaderboard are ultimately eliminated from the tournament.

- Each leaderboard of the next round will have a limited amount of available spots that is at least half of the total numnber of participants competing in the current round.

- Leaderboards will decrease in available spots for the next round whenever no one manages to secure a spot: in other words, after everyone competes in their current match, the leaderboard decreases in available spots when _NOT_ one competitor manages to win. The following demonstrates the amount of spots that are eliminated whenever a total stalemate occurs in the round.

### Performance Record

Competitors are placed on the leaderboard according to their performance record as indicated by the following:

Priority | Record
-|-
1 | Amount of Matches in Tournament
2 | Season Points (Total amount of points earned in the season, including points earned at the tournament)
3 | Fastest Submission in the Season, including the at tournament.

### Performance Comparison Examples for Tournaments:

#### Example 1

- Competitor A only has 3 matches

- Competitor B has 5 matches

- Competitor A has a better performance record: not as many attempts were made to advance in comparison to Competitor B.

#### Example 2

- Competitor A has earned a total of 4000 points for the entire season.

- Competitor B has earned a total of 2000 points for the entire season.

- Compatitor A earned 100 ponts while at the tournament, bringing the total to 4100 points.

- Competitor B earned 200 points while at the tournament, bringing the total to 2200 points.

- Competitor A has a better performance record: even though Competitor B is doing well at the tournament, Competitor A is doing better throughout the season.

#### Example 3

- Competitor A highest earned score from a match throughout the season is 250.

- Competitor B highest earned score from a match throughout the season is 200.

- Competitor A has a better performance record since their highest match score is more than the highest match score of Competitor B.

### Champion Title Hierarchy

Only one competitor will ultimately remain, which they will be awarded with a title of champion.

The performance record of the finalist will determine the champion title that they will receive.

Title | Guideline
-|-
Champion | Winner of tournament.
Undisputed Champion | Winner of World Championships and consistently had the best performance record throughout the entire season.
Supreme Champion | Winner of World Championships and earned a season points total of at least 4800.

## VII. Extreme Circumstances

It is possible for no one to advance to the next round, where the tournament roster begins with an even amount of participants and no one advances due to consistent stalemates.

If the amount of leaderboard spots happen to reduce to zero, then the participant with the best performance record will be crowned champion.

## X. Disqualifications

If the match ends in a Disqualification due to a rule violation, then the violator will lose the match.

The victim - participant who was violated - will receive points depending on the time when the referee ended the match.

For example: 

- 30 seconds have passed before _Competitor A_ did an illegal choke slam on opponent, _Competitor B_.
- The referee stopped the match because of this violation.
- _Competitor A_, the violator, ends up losing the match because of breaking the rules - diqualified.
- _Competitor B_, the one who was violated, wins the match and receives 270 points as appropriate compensation.

Only one warning may be given for the following _Prohibited Actions_: 1, 2, 3, 4, 5, 6, 7, 9.

Otherwise, any violation will result in an immediate disqualification of the current match.

**Two _Duel_ disqualifications will result in an elimination from the Circuit.**

**Two match disqualifications at a tournament will result in an elimination from the event, including _Qualifier Tournaments_ and _World Championships_.**
